```python
import pandas as pd
import yfinance as yf
from yahoofinancials import YahooFinancials
import sqlite3
import datetime
```

```python
etfs = pd.read_csv('etfs.csv')['Tickers'].tolist()

lengetfs = len(etfs)
counter = 0
etf_info = []
for etf in etfs:
    data = yf.Ticker(etf + ".SI")
    if data == None: 
        print("Error: " + etf)
    else:
        etf_info.append([data.info['shortName'], data.info['symbol']])
        counter += 1
        
print(f"Data for {counter} ETFs successfully retrieved out of {lengetfs}")
```

```python
all_data = {}
for etf in etfs:
    data = yf.download(etf+'.SI', start="2010-01-01", end="2023-06-30", period="1d")
    
    if data.empty:
        print(etf)
    else:
        all_data[etf] = data
        
print(len(all_data.keys()))
print(all_data['A35'])
```
